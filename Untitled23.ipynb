```python
import pandas as pd
df = pd.read_csv('test.csv')
```

```python
#การใช้  fiter()
df.filter(items=['Sex','Age']).query("Sex == 'female'")
```

```python
df.filter(items=['Sex','Age','Name','Fare']).query("Sex == 'female' and Age > 40")
```

```python
df.filter(items=['Name','Age','Pclass','Fare']).query("(Age.isnull() or Age >= 60) and Pclass == 1")
```

```python
df.loc[(df['Sex'] == 'female') & (df['Age'] > 30), ['Name', 'Sex', 'Age']]
```

```python
#loc
df.loc[
    (df['Sex']== 'female') &
    (df['Age'].notnull()) &
    (df['Pclass'].isin([1,2])),
    ['Name','Pclass','Sex','Age','Fare']
 ]
```

```python
df['mean_fare_by_class'] = df.groupby('Pclass')['Fare'].transform('mean')
df.loc[df['Fare'] > df['mean_fare_by_class']]
```

```python
df['mean_fare_by_class1'] = df.groupby('Pclass')['Fare'].transform('mean')
df.loc[df['Fare'] > df['mean_fare_by_class1']]
```

```python
df['Age_mean_by_class'] = df.groupby('Pclass')['Age'].transform('mean')

df.loc[
    (df['Sex'] == 'female') &
    (df['Age'] > df['Age_mean_by_class']),
    ['Name','Sex','Age', 'Pclass']
      ]
```

```python
df['Age_mean_by_class'] = df.groupby('Pclass')['Age'].transform('mean')
df['fare_by_class']     = df.groupby('Pclass')['Fare'].transform('mean')

df.loc[
    (df['Sex'] == 'female') &
    (df['Age'] > df['Age_mean_by_class']) &
    (df['Fare'] > df['fare_by_class']),
    ['Name','Sex','Age','Pclass', 'Fare']]
```

```python
df.assign(
    mean_fare_by_class = df.groupby('Pclass')['Fare'].transform('mean'),
    Age_by_class      = df.groupby('Pclass')['Age'].transform('mean')
).query(
    "'Fare' > 'mean_fare_by_class' and
    'Age' > 'fare_by_class'"
)[['Name','Sex','Age','Pclass','Fare']]
```

```python
df.assign(
    mean_fare_by_class = df.groupby('Pclass')['Fare'].transform('mean'),
    Age_by_class      = df.groupby('Pclass')['Age'].transform('mean')
).query(
    "Fare > mean_fare_by_clas and Age >         Age_by_class "
)[['Name','Sex','Age','Pclass','Fare']]
```

```python
df.assign(
    mean_fare_by_class = df.groupby('Pclass')['Fare'].transform('mean'),
    Age_by_class      = df.groupby('Pclass')['Age'].transform('mean')
).query(
    "Fare > mean_fare_by_class and Age >         Age_by_class "
)[['Name','Sex','Age','Pclass','Fare']]
```

```python
df.assign(
    mean_Age_by_class = df.groupby('Sex')['Age'].transform('mean')
).query(
    "Age > mean_Age_by_class"
)[['Name','Sex','Age','mean_Age_by_class']]
```
