```python
import pandas as pd

df = pd.read_csv('test.csv')
df_avarage = df['Age'].mean()
df_Fare    = df['Fare'].mean()
df['Age'] = df['Age'].fillna(df_avarage)
df['Fare'] = df['Fare'].fillna(df_Fare)
df.info()
#df.max[age]
#display(df.head(30))
#df_Max = df['Age'].max()
#df_std = df['Age'].std()
#df['Age'].fillna(df['df_Max'])
#print(f"ค่าเฉลี่ยคือ{df_avarage}")
#print(f"อายุสูงสุดคือ{df_Max}")
#print(f"ส่วนเบี่ยงเบนมาตราฐานคือ{df_std}")
```

```python
# 1. ศึกษาความสัมพันธ์ระหว่าง Pclass และ Fare
# - Pclass (ชั้นโดยสาร): 1, 2, 3
# - Fare (ค่าโดยสาร)

fare_by_class = df.groupby('Pclass')['Fare'].agg(['mean', 'median', 'min', 'max'])
print("\n--- ค่าโดยสารเฉลี่ย/มัธยฐาน ตามชั้นโดยสาร ---")
print(fare_by_class)

# 2. ศึกษาความสัมพันธ์ระหว่าง Sex และ Fare
# (แม้ไฟล์ test.csv ไม่มี Survived แต่เราสามารถวิเคราะห์ตัวแปรอื่นได้)
fare_by_sex = df.groupby('Sex')['Fare'].mean()
print("\n--- ค่าโดยสารเฉลี่ยตามเพศ ---")
print(fare_by_sex)

# 3. ศึกษาขนาดครอบครัว (Feature Engineering เบื้องต้น)
# SibSp: จำนวนพี่น้อง/คู่สมรส, Parch: จำนวนพ่อแม่/ลูก
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
print("\n--- ขนาดครอบครัวที่พบมากที่สุด 5 อันดับ ---")
print(df['FamilySize'].value_counts().head(5))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ตั้งค่าให้แสดงผลได้ชัดเจนขึ้น
sns.set_style("whitegrid")
plt.rcParams['font.size'] = 12

# 1. Histogram: ดูการแจกแจงของอายุ
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Passenger Age (Age)')
plt.xlabel('Age (Years)')
plt.show()

# 2. Box Plot: ดู Outliers และการกระจายตัวของค่าโดยสารตามชั้นโดยสาร
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', data=df, palette='pastel')
plt.title('Fare Distribution by Passenger Class (Pclass)')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
