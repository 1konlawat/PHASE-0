```python
import pandas as pd
import pandas as pd

df_customers = pd.DataFrame({
    'cust_id': [1, 2, 3, 4],
    'name': ['Aom', 'Best', 'Chin', 'Dew'],
    'city': ['Bangkok', 'Chiangmai', 'Bangkok', 'Khonkaen']
})
df_orders = pd.DataFrame({
    'order_id': [101, 102, 103, 104, 105],
    'cust_id': [1, 2, 2, 3, 5],  # มี cust_id ที่ไม่มีใน customers (5)
    'amount': [500, 1200, 800, 300, 999]
})
df_payments = pd.DataFrame({
    'order_id': [101, 102, 103, 106],  # มี order 106 ที่ไม่มีอยู่ใน orders
    'pay_status': ['paid', 'paid', 'unpaid', 'paid']
})
```

```python
df1 = df_customers.merge(df_orders, on='cust_id', how='left').merge(df_payments, on='order_id', how='left')
df2 = df1['amount'].fillna(0)
df1['amount'] =df2
df1
```
