```python
import pandas as pd

df1 = pd.DataFrame({
    "Drink": ["Latte","Latte","Americano","Americano","Tea","Tea"],
    "Size": ["M","L","M","L","M","L"],
    "Price": [70,80,60,70,40,50]
})
df2 =df1.groupby("Drink")["Price"].agg(['mean','max','min'])
df3 =df1.groupby('Size')['Price'].mean()
```

```python
import pandas as pd

df = pd.DataFrame({
    "Flavor":["Mint Chocolate Chip","Chocolate","Vanilla","Cookie Dough","Rocky Road","Pistachio","Cake Batter","Neapolitan","Chocolte Fudge Brownie"],
    "Base Flavor":["Vanilla","Chocolate","Vanilla","Vanilla","Chocolate","Vanilla","Vanilla","Vanilla","Chocolate"],
    "Liked":["Yes","Yes","No","Yes","Yes","No","Yes","No","Yes"],
    "Flavor Rating":[10,8.8,4.7,6.9,8.2,2.3,6.5,3.8,8.2],
    "Texture Rating":[8,7.6,5,6.5,7,3.4,6,5,7.1],
    "Total Rating":[18,16.6,9.7,13.4,15.2,5.7,12.5,8.8,15.3]
})
df1 = df.groupby('Base Flavor')['Total Rating'].mean()
df1
```

```python
df2 = df.groupby('Base Flavor')['Base Flavor'].count()
df2
```

```python
#การเปลี่ยนชื่อ
df3 = df.groupby('Base Flavor')['Total Rating'].agg(['min', 'max', 'mean']).rename(columns={'min':'น้อยที่สุด', 'max':'มากที่สุด', 'mean':'ค่าเฉลี่ย'})
df3
```
