```python
import pandas as pd

df2 = pd.DataFrame({
    'OrderID':[101,102,103,104,105,106,107,108],
    'Product':['A','A','B','B','C','A','C','B'],
    'Region':['North','South','North','West','South','North','West','South'],
    'Qty':[3,5,2,4,1,6,3,7],
    'Price':[100,100,200,200,300,100,300,200]
})
df2['Total'] = df2['Qty'] * df2['Price']
df2.groupby(['Region','Product'])['Total'].mean()
```

```python
import pandas as pd

sales = pd.DataFrame({
    'Branch' : ['A','A','B','B','A','C','C','B','C'],
    'Category' : ['Food','Drink','Food','Snack','Snack','Drink','Food','Drink','Snack'],
    'Qty' : [10,5,7,3,4,6,8,2,9],
    'Price' : [50,40,60,30,20,45,55,35,25]
})

sales['Total'] = sales['Qty'] * sales['Price']

sales.groupby(['Branch','Category'])['Total'].agg(['sum','mean']) \
.sort_values(by='sum', ascending=False)
```

```python
import pandas as pd

emp = pd.DataFrame({
    'Dept' : ['HR','IT','Finance','IT','HR','Finance','IT','HR','Finance'],
    'Gender' : ['F','M','F','F','M','M','M','F','F'],
    'Salary' : [30000,50000,45000,55000,27000,48000,60000,32000,52000],
    'Bonus' : [3000,5000,4000,6000,2000,4500,6500,2500,5000]
})

emp['TotalPay'] = emp['Salary'] + emp['Bonus']
emp.groupby(['Dept','Gender'])['TotalPay'].agg('mean')
```
